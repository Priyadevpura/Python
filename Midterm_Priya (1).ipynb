## Question 1:

You are working on developing a CRM system that prints vital information for a customer given certain variables. Write a Python code to perform the following tasks using f-strings by incorporating the variable names in the f-strings:



```
customer_id = "1234"
name = "Angela Martin"
home_latitude = 40.7128  # Latitude of Angela's home
home_longitude = -74.0060  # Longitude of Angela's home
current_latitude = 34.0522  # Latitude of Angela's current location
current_longitude = -118.2437  # Longitude of Angela's current location
```

1. Calculate the distance in miles from Angela's home to her current location and store it in a variable 'distance'.

2. Insert the 'distance' variable into the provided sentence using an f-string.

*Customer ID: 1234. Angela Martin's home coordinates are (40.7128, -74.0060), and her current coordinates are (34.0522, -118.2437). The distance from her home to her current location is ___ miles.*

You need to use the `geodesic` function from the `geopy.distance` library to calculate the miles using the provided latitude and longitude information.

````python
# your response here

from geopy.distance import geodesic

customer_id = "1234"
name = "Angela Martin"
home_latitude = 40.7128  # Latitude of Angela's home
home_longitude = -74.0060  # Longitude of Angela's home
current_latitude = 34.0522  # Latitude of Angela's current location
current_longitude = -118.2437  # Longitude of Angela's current location

# Calculate the distance in miles from Angela's home to her current location
home_coordinates = (home_latitude, home_longitude)
current_coordinates = (current_latitude, current_longitude)
distance = geodesic(home_coordinates, current_coordinates).miles

# Insert the 'distance' variable into the provided sentence using an f-string
sentence = f"Customer ID: {customer_id}. {name}'s home coordinates are {home_coordinates}, and her current coordinates are {current_coordinates}. The distance from her home to her current location is {distance:.2f} miles."

print(sentence)
````

## Question 2:

You are analyzing the performance of an email marketing campaign for a firm with 20 customers. The campaign data includes the following lists of values:



```
# Number of emails sent to 20 customers
emails_sent = [100, 120, 90, 150, 110, 130, 100, 140, 120, 110, 130, 150, 140, 120, 100, 110, 90, 130, 150, 140]

# Number of customers who clicked on the links in the emails
clicks = [10, 12, 8, 13, 11, 15, 10, 14, 12, 11, 13, 15, 14, 12, 10, 11, 9, 13, 15, 14]
```

Create a Python function called calculate_ctr that takes two parameters: emails_sent and clicks. Inside the function, compute the Customer Click-Through Rate (CTR) for each customer, which is the ratio of the number of customers who clicked on the links to the number of emails sent. Return the computed CTR.

Write a loop that iterates over the emails_sent and clicks lists for 20 customers and computes their click through rate using the calculate_ctr function. For each customer, calculate their CTR and append it to a new list called customer_ctr. Print the customer_ctr list containing the CTR for each customer.

After calculating the CTR for each customer, write Python statements to find and print the customer with the highest CTR and the customer with the lowest CTR from the customer_ctr list.


````python
# your answer here


emails_sent = [100, 120, 90, 150, 110, 130, 100, 140, 120, 110, 130, 150, 140, 120, 100, 110, 90, 130, 150, 140]

clicks = [10, 12, 8, 13, 11, 15, 10, 14, 12, 11, 13, 15, 14, 12, 10, 11, 9, 13, 15, 14]

def calculate_ctr(emails_sent, clicks):
    return [click / email_sent for email_sent, click in zip(emails_sent, clicks)]

customer_ctr = calculate_ctr(emails_sent, clicks)

for customer, ctr in enumerate(customer_ctr, start=1):
    print(f"Customer {customer}: CTR = {ctr:.2%}")

max_ctr_customer = customer_ctr.index(max(customer_ctr)) + 1
print(f"Customer {max_ctr_customer} has the highest CTR: {max(customer_ctr):.2%}")


min_ctr_customer = customer_ctr.index(min(customer_ctr)) + 1
print(f"Customer {min_ctr_customer} has the lowest CTR: {min(customer_ctr):.2%}")
````

## Question 3:

You are analyzing the performance of multiple companies operating in the market during a specific quarter. The market data for that quarter includes the following lists and data for different companies:



```
# Revenue values (in dollars) for several companies in the market during a specific quarter
company_revenues = [50000, 52000, 55000, 48000, 51000, 48000, 49000, 52000, 51000, 53000]

# Names of the companies
company_names = ["Company A", "Company B", "Company C", "Company D", "Company E", "Company F", "Company G", "Company H", "Company I", "Company J"]

# Total market revenue (in dollars) for that specific quarter
total_market_revenue = 250000

```

Create a Python function called calculate_market_share that takes two input parameters: company_revenues, and total_market_revenue. Inside the function, compute the Market Share for each company based on its revenue and the total market revenue. Return a list (mkt_share) that captures the market share for each company.

Next, loop over company_names and mkt_share and return a dictionary where each company name is the key, and the corresponding Market Share is the value. Print the market_share_dict containing the Market Share for each company.


````python

company_revenues = [50000, 52000, 55000, 48000, 51000, 48000, 49000, 52000, 51000, 53000]
company_names = ["Company A", "Company B", "Company C", "Company D", "Company E", "Company F", "Company G", "Company H", "Company I", "Company J"]


total_market_revenue = 250000

def calculate_market_share(company_revenues, total_market_revenue):
    return [(revenue / total_market_revenue) for revenue in company_revenues]


mkt_share = calculate_market_share(company_revenues, total_market_revenue)

market_share_dict = dict(zip(company_names, mkt_share))

for company, market_share in market_share_dict.items():
    print(f"{company}: Market Share = {market_share:.2%}")
````

````python
from plotnine import ggplot, aes, geom_point, geom_smooth, labs, theme_minimal

# Create a scatterplot to visualize the relationship between age and daily time spent
scatterplot_age_time = ggplot(clickad_data, aes(x='Age', y='Daily Time Spent on Site')) + \
    geom_point(alpha=0.6) + \
    geom_smooth(method='lm', se=False, color='blue') + \
    labs(title="Relationship between Age and Daily Time Spent on Site", x="Age", y="Daily Time Spent on Site") + \
    theme_minimal()

scatterplot_age_time.draw()
````

This scatterplot will help us to visualize the relationship between age and daily time spent without considering gender.

````python
# Create a scatterplot with separate regression lines for males and females
scatterplot_age_time_gender = ggplot(clickad_data, aes(x='Age', y='Daily Time Spent on Site', color='factor(Male)')) + \
    geom_point(alpha=0.6) + \
    geom_smooth(method='lm', se=False) + \
    labs(title="Relationship between Age, Daily Time Spent, and Gender", x="Age", y="Daily Time Spent on Site") + \
    theme_minimal()
scatterplot_age_time_gender.draw()
````

Use the same dataset as in question 7 to answer the following question

1. What is the relationship between ad click and daily time spent? Explain using a suitable plot.
2. Does this relationship change in any way when you take into account their gender? Explain by modifying the above plot.

````python
from plotnine import ggplot, aes, geom_point, geom_smooth, facet_wrap, labs, theme_minimal

# Convert 'Clicked on Ad' to a numeric column (0 or 1)
clickad_data['Ad_Clicked'] = clickad_data['Clicked on Ad'].astype(int)

# Create a scatterplot with regression lines for ad click and daily time spent
scatterplot_ad_click_time = (
    ggplot(clickad_data, aes(x='Daily Time Spent on Site', y='Ad_Clicked')) +
    geom_point(alpha=0.6) +
    geom_smooth(method='lm', se=False, color='blue') +
    labs(title="Relationship between Ad Click and Daily Time Spent", x="Daily Time Spent on Site", y="Clicked on Ad") +
    theme_minimal()
)

# Create a scatterplot with separate regression lines for males and females
scatterplot_ad_click_time_gender = (
    ggplot(clickad_data, aes(x='Daily Time Spent on Site', y='Ad_Clicked', color='factor(Male)')) +
    geom_point(alpha=0.6) +
    geom_smooth(method='lm', se=False) +
    facet_wrap('~Male', ncol=1) +
    labs(title="Relationship between Ad Click, Daily Time Spent, and Gender", x="Daily Time Spent on Site", y="Clicked on Ad") +
    theme_minimal()
)

print(scatterplot_ad_click_time)
print(scatterplot_ad_click_time_gender)
````

1. In the first plot, we see overall relationship betweem ad click and daily time spend, with a single regression line.
